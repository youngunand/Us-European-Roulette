# Us-European Roulette Comparison

This file will compare US-Roulette to European Roulette by simulating the same strategies on the different wheels.
While there are two zeros in US-Roulette (0 and 00), there is only one in the European Version of Roulette. Therefore the odds and outcomes are changing.

## Simulation

Method of Simulation: This simulation is inspired by the Youtube-Channel Minding The Data. In a Video he simulates the outcomes of different strategies in Us-Roulette using Python. For the comparison I used and modified some of his algortihms and added new ones myself. 

The libraries matplotlib and seaborn are used to show the data.

## Martingale Strategy

The Martingale is one of the most popular strategies for Roulette. The player always doubles his bet with a loss. It provides the player with long and steady profits, but can end suddenly every time since the betsize increases exponantially.

### Simulating single streaks

```python
#Martingale-Strategy: always double your bet until you lose (long steady profits, but suddenly ending)
#European Roulette
def martingale_european(bankroll):
    bet = 0.01
    pockets = ["Red"] * 18 + ["Black"] * 18 + ["Green"] 
    bankroll_history = []
    while bankroll > 0: 
        if bet > bankroll:
            bet = bankroll
        roll = random.choice(pockets)
        if roll == "Red":
            bankroll += bet
            bet = 0.01
        else:
            bankroll -= bet
            bet *=2
        bankroll_history.append(bankroll)
    return bankroll_history
```

As you can see, we only add one green pocket to the array of pockets. That results in a chance of 18/37 of hitting Red and a 19/37 chance of not hitting red (green or black).
We assume that the player will always stay with red, but it wouldn't make any difference if he chose black.
The first bet of the player will be 0.01 Euro, which means every time he wins he will take 0.01 Euros of profit. As long as he doesn't go bankrupt, he will make 0.01 Euro with every betting streak (He wins back his loss and the starting bet remains as profit).
However, one longer streak of bad luck makes the player go bankrupt.
The bet size is that small, because this way the player can survive longer and we get more clear and statistically correct results.

```python
#Martingale-Strategy: always double your bet until you lose (long steady profits, but suddenly ending)
#Us-American Roulette
def martingale_us(bankroll):
    bet = 0.01
    pockets = ["Red"] * 18 + ["Black"] * 18 + ["Green"] * 2
    bankroll_history = []
    while bankroll > 0: 
        if bet > bankroll:
            bet = bankroll
        roll = random.choice(pockets)
        if roll == "Red":
            bankroll += bet
            bet = 0.01
        else:
            bankroll -= bet
            bet *=2
        bankroll_history.append(bankroll)
    return bankroll_history
```

The function for the Us-Simulation is the same, except that there are 2 green pockets added to the array instead of one. This changes the probability of hitting red to 18/38 and the probability of not hitting red (hitting black or green) to 20/38.

We simulate the strategy 20 times and set up seaborn and matplotlib in order to get a nice graph. The player will start with a bankroll of 100.

```python
%matplotlib inline
import random
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(13.7,8.27)})

for i in range(20):
    plt.plot(martingale_european(bankroll=100), linewidth=2)
    
    
plt.xlabel("Number of Games", fontsize=18, fontweight="bold")
plt.ylabel("Bankroll", fontsize=18, fontweight="bold")
plt.xticks(fontsize=16, fontweight="bold")
plt.yticks(fontsize=16, fontweight="bold")
plt.title("European Martingale", fontsize=22, fontweight="bold")
plt.show()
```

```python
%matplotlib inline
import random
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(13.7,8.27)})

for i in range(20):
    plt.plot(martingale_us(bankroll=100), linewidth=2)
    
    
plt.xlabel("Number of Games", fontsize=18, fontweight="bold")
plt.ylabel("Bankroll", fontsize=18, fontweight="bold")
plt.xticks(fontsize=16, fontweight="bold")
plt.yticks(fontsize=16, fontweight="bold")
plt.title("Us-American Martingale", fontsize=22, fontweight="bold")
plt.show()
```

We can see that the chances of the player are much better playing European-Roulette. He can reach a longer streak and can also reach a higher bankroll much faster.
